```
import pandas as pd
dataSet = pd.read_csv('zomato.csv', encoding='ISO-8859-1')
dataSet.head(5)
```

Created a function called findVal and Added the Regex

which search following keywords 

Keywords = ["firstfloor","first floor", "1st floor", "1stfloor", "1floor", "secondfloor", "second floor","2nd floor", 1st flr, 1stflr]



```
import re
def findVal(value):
    if(re.search("((1|2)(st|nd)[ ]*fl[o]*r)|((first|second)[ ]*fl[o]*r)", value.lower()) != None):
        return 1
    else:
        return 0    
dataSet["FloorData"] = dataSet["Address"].apply(findVal)
dataSet[dataSet["FloorData"]==1][["Restaurant Name", "Address"]]

# This tells we have 525 Restaurants that are there in first floor and 2nd floor
```
